```python
import numpy as np
import pandas as pd
train_set = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header = None)
test_set = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test" , skiprows = 1, header = None)
col_labels = ["age", "workclass", "fnlwgt", "education","education_num","marital_status", "occupation","relationship", "race", "sex", "capital_gain",
"capital_loss", "hours_per_week", "native_country", "wage_class"]
train_set.columns = col_labels
test_set.columns = col_labels
```

```python
df = pd.concat([test_set,train_set],axis=0)
```

```python
df["workclass"].replace({' ?':df.workclass.mode().iloc[0]},inplace=True)
df.occupation.replace({' ?':df.occupation.mode().iloc[0]},inplace=True)
df.native_country.replace({' ?':df.native_country.mode().iloc[0]},inplace=True)
```

```python
df["wage_class"].replace({' <=50K.':' <=50K'},inplace=True)
df["wage_class"].replace({' >50K.':' >50K'},inplace=True)
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
categorical_features = [i for i in x.columns if df[i].dtype=="object"]
```

```python
new_x_dummies = pd.get_dummies(x[categorical_features],drop_first=True)
```

```python
new_x_dummies = pd.concat([x,new_x_dummies],axis=1)
```

```python
new_x_dummies.drop(categorical_features,axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder(drop='first',dtype=int)
encoded_frame = pd.DataFrame((one_hot_encoder.fit_transform(x[categorical_features])).toarray())
```

```python
new_x_one = x.join(encoded_frame)
```

```python
new_x_one.drop(categorical_features,axis=1,inplace=True)
```
